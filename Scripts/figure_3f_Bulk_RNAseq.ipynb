````python
data = pd.read_csv("/home/q5gong/CSE185FinalProject/Figure_3f/data/DESeq2_Result")
data = data.rename(columns={"Unnamed: 0": "gene_id"})
data.head()
````

````python
expression = data.drop("gene_id", axis="columns")
expression.head()
````

````python
cg = sns.clustermap(expression.transpose(), z_score=1, cmap="coolwarm")
cg.ax_row_dendrogram.set_visible(False)
cg.ax_col_dendrogram.set_visible(False)
````
