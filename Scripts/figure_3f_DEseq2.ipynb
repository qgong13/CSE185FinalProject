```r
# Load the libraries
library("DESeq2")
library("tximport")
```

```r
data <- read.csv("/home/q5gong/CSE185FinalProject/Figure_3f/data/GSE132329_Alonso-Curbelo_RNASeq_SET2.csv",
                 row.name="ID")
head(data)
names(data)
```

```r
countdata <- subset(data, select=c(Normal.shRen.713_RNA1, Normal.shRen.713_RNA2, Normal.shRen.713_RNA3, 
                                   RegADM.shRen.713_RNA1, RegADM.shRen.713_RNA2, RegADM.shRen.713_RNA3,
                                   RegADM.shBrd4.1448_RNA1, RegADM.shBrd4.1448_RNA2, RegADM.shBrd4.1448_RNA3,
                                   Kras.ADR.shRen.713_RNA1, Kras.ADR.shRen.713_RNA2, Kras.ADR.shRen.713_RNA3, 
                                   Kras.ADR.shBrd4.1448_RNA1, Kras.ADR.shBrd4.1448_RNA2, Kras.ADR.shBrd4.1448_RNA3))
head(countdata)
names(countdata)
```

```r
cts <- as.matrix(countdata)
conditions <- c(rep("Normal_CshRen", 3),
                rep("Injury_CshRen", 3),
                rep("Injury_CshBrd4", 3),
                rep("Kras*+Injury_CshRen", 3),
                rep("Kras*+Injury_CshBrd4", 3))
samples <- data.frame("run"=c("Normal_CshRen.713_RNA1", "Normal_CshRen.713_RNA2", "Normal_CshRen.713_RNA3", 
                              "Injury_CshRen.713_RNA1", "Injury_CshRen.713_RNA2", "Injury_CshRen.713_RNA3",
                              "Injury_CshBrd4.1448_RNA1", "Injury_CshBrd4.1448_RNA2", "Injury_CshBrd4.1448_RNA3", 
                              "Kras*+Injury_KCshRen.713_RNA1", "Kras*+Injury_KCshRen.713_RNA2", "Kras*+Injury_KCshRen.713_RNA3",
                              "Kras*+Injury_KCshBrd4.1448_RNA1", "Kras*+Injury_KCshBrd4.1448_RNA2", "Kras*+Injury_KCshBrd4.1448_RNA3"),
                      "condition"=conditions)
```

```r
dds <- DESeqDataSetFromMatrix(countData = cts, 
                              colData = samples, 
                              design = ~condition)
dds <- DESeq(dds)
dds
```

```r
# Normal_CshRen v.s. Injury_CshRen
res1 <- results(dds, contrast=c("condition", "Normal_CshRen", "Injury_CshRen"))
res1
```

```r
# Normal_CshRen v.s. Injury_CshBrd4
res2 <- results(dds, contrast=c("condition", "Normal_CshRen", "Injury_CshBrd4"))
res2
```

```r
# Normal_CshRen v.s. Kras*+Injury_CshRen
res3 <- results(dds, contrast=c("condition", "Normal_CshRen", "Kras*+Injury_CshRen"))
res3
```

```r
# Normal_CshRen v.s. Kras*+Injury_CshBrd4
res4 <- results(dds, contrast=c("condition", "Normal_CshRen", "Kras*+Injury_CshRen"))
res4
```

```r
# find the significant gene
sigGenes <- unique(c(
  rownames(subset(res1, padj < 0.05 & abs(log2FoldChange) > 2)),
  rownames(subset(res2, padj < 0.05 & abs(log2FoldChange) > 2)),
  rownames(subset(res3, padj < 0.05 & abs(log2FoldChange) > 2)),
  rownames(subset(res4, padj < 0.05 & abs(log2FoldChange) > 2))
))
```

```r
final_res <- assay(dds)[sigGenes,]
final_res
```

```r
write.csv(as.data.frame(final_res),
          file="/home/q5gong/CSE185FinalProject/Figure_3f/data/DESeq2_Result")
```
