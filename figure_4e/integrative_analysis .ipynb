### Due to time, instead of doing the integrative analysis, just use the extended table 5

````python
integrative = pd.read_excel("data/41586_2020_3147_MOESM7_ESM.xlsx",
                            sheet_name="KrasG12DInjury_vs_Normal", index_col=0)
integrative.dropna(inplace=True)
````

````python
integrative.head()
````

````python
integrative.isnull().any()
````

````python
integrative_region = pd.merge(left=integrative, right=RNA_region, how='left',
                              left_on='ID', right_on='ID')
integrative_region.head()
````

````python
integrative_region.isnull().any()
````

````python
integrative_region_bed = {"chrom": [],
                          "start_position": [], 
                          "end_position": []}

for chrom, start, end in zip(integrative_region.chromosome_name,
                             integrative_region.start_position,
                             integrative_region.end_position):
    
    integrative_region_bed["chrom"].append("chr" + str(chrom))
    integrative_region_bed["start_position"].append(start)
    integrative_region_bed["end_position"].append(end)

integrative_region_BED_df = pd.DataFrame(integrative_region_bed)
integrative_region_BED_df = integrative_region_BED_df.set_index("chrom")
integrative_region_BED_df.head()
````

````python
# create a pseudo bed file based on integrate information
integrative_region_BED_df.to_csv('data/integrative.bed', sep="\t")
````

### The pseudo bed for scATACseq

````python
import scanpy as scanpy
````

````python
adata = scanpy.read_10x_h5("data/scATAC-seq/GSE137069_Merged_filtered_peak_bc_matrix.h5",
                           genome=None, gex_only=False)
````

````python
scATAC = pd.DataFrame(adata.var)
scATAC.head()
````

````python
scATAC_bed = {"chrom": [],
              "start_position":[],
              "end_position":[]}

for gene_ids in scATAC.gene_ids:
    chrom_region = gene_ids.split(":")
    chrom = chrom_region[0]
    region = chrom_region[1]
    
    start_end = region.split("-")
    start = start_end[0]
    end = start_end[1]
    
    scATAC_bed["chrom"].append(chrom)
    scATAC_bed["start_position"].append(start)
    scATAC_bed["end_position"].append(end)
    
scATAC_BED_df = pd.DataFrame(scATAC_bed)
scATAC_BED_df = scATAC_BED_df.set_index("chrom")
scATAC_BED_df.head()    
````

````python
# create a pseudo bed file based on scATACseq information
scATAC_BED_df.to_csv('data/scATAC.bed', sep="\t")
````
