```r
countdata <- subset(data, select=c(Normal.shRen.713_RNA1, Normal.shRen.713_RNA2, Normal.shRen.713_RNA3, 
                                   RegADM.shRen.713_RNA1, RegADM.shRen.713_RNA2, RegADM.shRen.713_RNA3,
                                   RegADM.shBrd4.1448_RNA1, RegADM.shBrd4.1448_RNA2, RegADM.shBrd4.1448_RNA3,
                                   Kras.ADR.shRen.713_RNA1, Kras.ADR.shRen.713_RNA2, Kras.ADR.shRen.713_RNA3, 
                                   Kras.ADR.shBrd4.1448_RNA1, Kras.ADR.shBrd4.1448_RNA2, Kras.ADR.shBrd4.1448_RNA3,
                                   Kras.ADR.shBrd4.552_RNA1, Kras.ADR.shBrd4.552_RNA2, Kras.ADR.shBrd4.552_RNA3))
head(countdata)
names(countdata)
```

```r
cts <- as.matrix(countdata)
conditions <- c(rep("Normal_CshRen", 3),
                rep("Injury_CshRen", 3),
                rep("Injury_CshBrd4", 3),
                rep("Kras*+Injury_CshRen", 3),
                rep("Kras*+Injury_CshBrd4", 3),
                rep("Kras*+Injury_CshBrd4_552", 3))
samples <- data.frame("run"=c("Normal_CshRen.713_RNA1", "Normal_CshRen.713_RNA2", "Normal_CshRen.713_RNA3", 
                              "Injury_CshRen.713_RNA1", "Injury_CshRen.713_RNA2", "Injury_CshRen.713_RNA3",
                              "Injury_CshBrd4.1448_RNA1", "Injury_CshBrd4.1448_RNA2", "Injury_CshBrd4.1448_RNA3", 
                              "Kras*+Injury_KCshRen.713_RNA1", "Kras*+Injury_KCshRen.713_RNA2", "Kras*+Injury_KCshRen.713_RNA3",
                              "Kras*+Injury_KCshBrd4.1448_RNA1", "Kras*+Injury_KCshBrd4.1448_RNA2", "Kras*+Injury_KCshBrd4.1448_RNA3",
                              "Kras.ADR.shBrd4.552_RNA1", "Kras.ADR.shBrd4.552_RNA2", "Kras.ADR.shBrd4.552_RNA3"),
                      "condition"=conditions)
```

```r
dds <- DESeqDataSetFromMatrix(countData = cts, 
                              colData = samples, 
                              design = ~condition)
dds <- DESeq(dds)
dds
```

### Condition:
The C-sh group is the normal mice group and the KC-sh group is the mice group with Kras* mutation. The shRen and shBrd4 represent suppresion of the Rennila luciferace (control) and BRD4. 

Hence, check differtail expression between the following group and output the significantly expressed gene based on the comparsion result:
1. Normal_CshRen v.s. Injury_CshRen
2. Normal_CshRen v.s. KrasInjury_CshRen
3. Normal_CshRen v.s. Injury_CshBrd4
4. Normal_CshRen v.s. KrasInjury_CshBrd4

```r
# Normal_CshRen v.s. Injury_CshRen
res1 <- results(dds, contrast=c("condition", "Normal_CshRen", "Injury_CshRen"))
```

```r
# Normal_CshRen v.s. Kras*+Injury_CshRen
res2 <- results(dds, contrast=c("condition", "Normal_CshRen", "Kras*+Injury_CshRen"))
```

```r
# Normal_CshRen v.s. Injury_CshBrd4
res3 <- results(dds, contrast=c("condition", "Normal_CshRen", "Injury_CshBrd4"))
```

```r
# Normal_CshRen v.s. Kras*+Injury_CshBrd4
res4 <- results(dds, contrast=c("condition", "Normal_CshRen", "Kras*+Injury_CshBrd4"))
```

```r
# Injury_CshRen v.s. Injury_CshBrd4
res5 <- results(dds, contrast=c("condition", "Injury_CshBrd4", "Injury_CshRen"))
regeneration_all <- subset(res5, padj < 0.05 & log2FoldChange < -2)
regeneration_sorted <- regeneration_all[order(regeneration_all$padj),]
regeneration <- rownames(regeneration_sorted)
```

```r
# Kras*+Injury_CshRen v.s. Kras*+Injury_CshBrd4
res6 <- results(dds, contrast=c("condition", "Kras*+Injury_CshBrd4", "Kras*+Injury_CshRen"))
neoplasia_all <- subset(res6, padj < 0.05 & log2FoldChange < -2)
neoplasia_sorted <- neoplasia_all[order(neoplasia_all$padj),]
neoplasia <- rownames(neoplasia_sorted)
```

```r
# find the significant gene
sigGenes <- unique(c(
#   rownames(subset(res1, padj < 0.05 & abs(log2FoldChange) > 2)),
#   rownames(subset(res2, padj < 0.05 & abs(log2FoldChange) > 2)),
#   rownames(subset(res3, padj < 0.05 & abs(log2FoldChange) > 2)),
#   rownames(subset(res4, padj < 0.05 & abs(log2FoldChange) > 2)),
  rownames(subset(res5, padj < 0.05 & log2FoldChange < -2)),
  rownames(subset(res6, padj < 0.05 & log2FoldChange < -2))
#     rownames(res5), rownames(res6)
))
```

```r
final_res <- assay(dds)[sigGenes,]
```

```r
write.csv(as.data.frame(final_res),
          file="/home/q5gong/CSE185FinalProject/Figure_3f/data/DESeq2_Result.csv")
```

### Visualization with pheatmap

```r
specific_genes <- unique(c(regeneration, neoplasia))
```

```r
temp_df <- as.data.frame(final_res)
# common <- temp_df %>% filter(!row.names(temp_df) %in% specific_genes)
# common <- rownames(common)
common <- intersect(regeneration, neoplasia)
neoplasia_specific <- setdiff(neoplasia, common)
regeneration_specific <- setdiff(regeneration, common)
```

```r
anno_order <- c(regeneration_specific, common, neoplasia_specific)
temp_df <- temp_df[anno_order, ]
write.csv(temp_df,
          file="/home/q5gong/CSE185FinalProject/Figure_3f/data/DESeq2_Ordered_Result.csv")
```

```r
write.csv(as.data.frame(regeneration_specific),
          file="/home/q5gong/CSE185FinalProject/Figure_3f/data/regeneration_specifc.csv")
write.csv(as.data.frame(neoplasia_specific),
          file="/home/q5gong/CSE185FinalProject/Figure_3f/data/neoplasia_specifc.csv")
write.csv(as.data.frame(common),
          file="/home/q5gong/CSE185FinalProject/Figure_3f/data/common.csv")
```

```r
Z <- scale(t(final_res))
Z <- Z[-c(16, 17, 18),]
Z <- Z[, anno_order, drop=FALSE]

pheatmap(Z, color=colorRampPalette(c("#85C1E9", "white", "#FF5757"))(17),
         cluster_rows=FALSE, show_colnames=FALSE, cluster_cols=FALSE)
```

```r
# Change Jupyter plots size
options(repr.plot.width=14, repr.plot.height=6)

row_category = c(rep("1Normal_shRen", 3),
                 rep("2Injury_shRen", 3),
                 rep("3Injury_shBrd4", 3),
                 rep("4Kras_shRen", 3),
                 rep("5Kras_shBrd4", 3))
column_category = c(rep("1Regeneration", length(regeneration_specific)),
                    rep("2Common", length(common)),
                    rep("3Neoplasia", length(neoplasia_specific)))

anno_color = c("#DAF7A6", "#BBBBBB", "#B83838")
palette = colorRamp2(c(-2, 0, 2), c("#85C1E9", "white", "#FF5757"))

#png(file="/home/q5gong/CSE185FinalProject/Figure_3f/figure3f.png")
ht <- Heatmap(Z, name="z-score",
              row_split=row_category, row_order=rownames(Z),
              row_title=c("Normal_shRen", "Injury_shRen", "Inury_shBrd4",
                          "Kras_shRen", "Kras_shBrd4"), row_title_rot=0,
              column_split=column_category, column_order=anno_order,
              column_title=c("Regeneraion", "Common", "Neoplasia"),
              column_title_rot=0, col=palette,
              show_row_names=FALSE, show_column_names=FALSE, 
              width=unit(12, "in"), height=unit(4, "in"),
              top_annotation = HeatmapAnnotation(foo = anno_block(gp = gpar(fill = anno_color))))
draw(ht)
#dev.off()
```
