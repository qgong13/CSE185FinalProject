### Paper Summary:
KRASÂ mutation is found to be associated with pancreatic ductal adenocarcinoma (PDAC), a type of damage-associated carcinogenesis that is lethal and, unfortunately, lacks effective therapies. In their search for the mechanism of mutantÂ KRASÂ induced PDAC, Alonso-Curbelo et al. revealed that a genetic change (the mutatedÂ KRAS) in combination with environmental change (tissue damage) promotes an epigenetic reprogramming that involved acinar-to-neoplasia chromatin switch in the pancreatic cells.Â 

The mechanism is illustrated in the figure below, with highlights listed in the following bullet points:
   >    
   >* WithÂ KRASÂ mutation, injured pancreatic cells gain a PDAC-like chromatin profile, which results in neoplasia-specific transcription.
   >* Interleukin-33 (IL-33) is the most robustly activated factor.
   >* With protein BRD4 that can access open chromatin, which is supposed to aid in re-establish of acinar-like chromatin profile, now aids in the increased expression of IL-33.
   >* IL-33 further promote neoplasia formation, which can then develop into PDAC

Overall, this study demonstrates how gene-environment interactions can alter gene-regulatory programs to initiate tumor formation, which opens up the possibility for novel therapeutic strategies to prevent (e.g., inhibit expression of BRD4 to prevent the expression of IL-33) and detect (e.g., the target expression level of IL-33) tumors before they evolve to an advanced stage.
